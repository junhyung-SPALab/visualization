```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
aligned_t_1 = np.load('./aligned_features_2d/val_1__t_1.npy')
aligned_t_2 = np.load('./aligned_features_2d/val_1__t_2.npy')
spatial_t_0 = np.load('./spatial_features_2d/val_1__t_0.npy')
```

```python
data = np.concatenate([spatial_t_0, aligned_t_1, aligned_t_2])
```

aligned_t_1_ = aligned_t_1[0,:,:]
print(aligned_t_1_.shape)
print(np.max(aligned_t_1_))
print(np.min(aligned_t_1_))

```python
for data_ in data:
    for i in range(data_.shape[0]):
        f_min = np.min(data_[i])
        f_max = np.max(data_[i])
        data_[i]  = (data_[i] - f_min) / (f_max - f_min)
```

for i in range(aligned_t_1.shape[0]):
